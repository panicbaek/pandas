```python
# 결혼한 남자들의 사용금액 평균
a = df.query("성별 == 'M' and 기혼 == 'Married' ")['사용금액'].mean()
# 결혼 안한 남자들의 사용금액 평균
b = df.query(" 성별 == 'M' and 기혼 == 'Single' ")['사용금액'].mean()
print(f"결혼한 남자들 사용금액 평균 : {a}")
print(f"결혼 안한 남자들 사용금액 평균 : {b}")
print(f"결혼한 사람이 안한 사람들보다 : {a-b} ")
print(f"기혼 남성이 미혼 남성들보다 6% 더 많이 지출함 : {(a-b) / a * 100}")

```
