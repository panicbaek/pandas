```python
import seaborn as sns
import numpy as np
df = sns.load_dataset('tips')
df.head()
```

```python
# total_bill, tip, size의 평균과 표준편차구하기
df[ [ 'total_bill', 'tip', 'size'] ].agg(['mean', 'std'])
# tip의 최대값과 최소값
df['tip'].agg(['min', 'max'])

df['tip_rate'] = df['tip'] / df['total_bill'] * 100 # 팁이 지불금액에 몇 %인지

df.groupby('day')['tip'].mean() # 요일별 팁 평균

df.groupby('sex')['tip_rate'].mean() # 성별별 팁 비율 평균구하기

# tip_rate가 20%이상인 데이터 구하기
df[ df['tip_rate'] >=0.2 ] # df.query(("tip_rate >= 0.2")

# total_bill이 전체 평균보다 큰 데이터 구하기
df[ df['total_bill'] > df['total_bill'].mean() ]

# size가 4이상인 팁 평균 구하기
df[ df['size'] >= 4 ]['tip'].mean()

# time별 total_bill 평균, 합계 구하기
df.groupby('time')['total_bill'].agg(['mean', 'sum'])

# 성별과 시간대별 tip_rate 평균 구하기
df.groupby(['sex', 'time'])['tip_rate'].mean()
```

```python
# 알바를 하려고 하는데 언제 알바하는게 팁을 가장 많이 받을까?
df.groupby('time')['tip_rate'].agg(['mean', 'count'])
df.groupby(['time', 'day', 'sex', 'smoker', 'size']).agg(mean=("tip_rate", "mean"), n = ("tip_rate", "count"))\
  .query("n >= 10") \
  .sort_values('mean', ascending=False)
```
