```python
# 컬럼명 변경하기
# mpg의 cty => city, hwy => highway로 변경
mpg_rename = mpg.copy() # 복사본 생성
mpg_rename.rename(columns={'cty':'city', 'hwy':'highway'}, inplace=True)
mpg_rename
```

```python
# 컬럼 추가하기 ( 파생 변수 생성 )
# 기존 컬럼(변수)를 이용해서 만들어진 새 컬럼을 파생변수라고 함
df = pd.DataFrame({
  'var1' : [4,3,6,],
  'var2' : [2,6,1]
})
df
```

```python
df['sum'] = df['var1'] + df['var2']
df
```

```python
df['mean'] = df['sum']/2
df
```

```python
# mpg데이터 프레임에서 cty와 hwy의 평균값을 나타내는 
# total컬럼을 추가
mpg['total'] = mpg['cty'] + mpg['hwy'] /2
mpg
```

```python
# total이 20이상이면 pass, 아니면 fail
# 넘파이에 where이라는 함수를 이용해서 처리가 가능함
import numpy as np
mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail')
mpg.head()
```

```python
# 해당 컬럼에 값이 종류별 몇개씩 있는지 
mpg['test'].value_counts()
```

```python
# 조건을 여러개
# total이 30이상이면 A, 20이상이면 B, 나머지는 C
mpg['grade'] = np.where(mpg['total'] >= 30, 'A', np.where( mpg['total']>=20, 'B', 'C'))
mpg
```

```python
mpg['grade'].value_counts()
```

```python
mpg['category'].value_counts()
```

```python
mpg['size'] = np.where(
  (mpg['category'] == 'compact') |
  (mpg['category'] == 'subcompact') |
  (mpg['category'] == '2seater'),
  'small', 'large'
)

mpg['size'] = np.where(
  mpg['category'].isin(['compact', 'subcompact', '2seater']),
  'small', 'large'
)

mpg['size'].value_counts()

mpg
```

```python
# midwest.csv파일을 가져와서 데이터 구조 파악

# poptotal(전체인구)를 total로, 
# popasian(아시아인구)를 asian으로 변경

midwest = pd.read_csv('data/midwest.csv')
midwest_rename = midwest.copy()
midwest_rename.rename(columns={'poptotal' : 'total', 'popasian' : 'asian'}, inplace=True)
midwest_rename
```

```python
# totla과 asian을 이용해서 전체인구 대비 아시아 인구 비율
# 파생변수(ratio) 추가
midwest_rename['ratio'] = midwest_rename['asian'] / midwest_rename['total'] * 100
midwest_rename
```
