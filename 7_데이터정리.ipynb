데이터를 직접 만들어서 사용을 한다면 별 문제가 없음
분석하기 적합한 상태로 정리를 해야 함
-각 열들은 하나만 있어야 함
-행데이터는 중복되는 행이 있으면 안됨
-한칸에는 하나의 값만 있어야 함
-데이터 타입도 중요함
-결측치(값이 없는거)는 적어야 함

```python
import pandas as pd
df = pd.read_csv('data/pew.csv')
df.head(3)
```

```python
# 피벗 되돌리기 ( 초기화 )
df_long = df.melt(id_vars='religion', var_name='income', value_name='count')
df_long
```

```python
billboard = pd.read_csv('data/billboard.csv')
billboard.head(2)
```

```python
billboard_melt = billboard.melt(
  id_vars=['year', 'artist',	'track',	'time',	'date.entered'],
  var_name='week',
  value_name='rating'
)
billboard_melt
```

```python
ebola = pd.read_csv('data/country_timeseries.csv')
ebola.head(3)
```

```python
ebola_melt = ebola.melt(id_vars=['Date', 'Day'])
ebola_melt.head(3)
```

```python
var_split = ebola_melt.variable.str.split('_')
var_split
```

```python
country = var_split.str.get(1)
country
```
