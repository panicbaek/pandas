```python
import pandas as pd
df = pd.read_excel('data/product.xlsx') # 엑셀을 사용하고 싶으면 설치를 해야함 
# pip install openpyxl
df.columns
```

```python
# 부가세포함된 가격 컬럼을 추가
# 소비자가 파생변수를 추가 : 판매가 10%를 추가한 금액
def 소비자가(판매가):
  return 판매가 * 1.1

df['소비자가'] = df['판매가'].apply(소비자가)
df  
```

```python
import numpy as np
def get_category(name):
  if 'Chair' in str (name):
    return '의자'
  
df['카테고리'] = df['상품목록'].apply(get_category)
df
```
